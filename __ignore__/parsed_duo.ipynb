```python
import pandas as pd
import numpy as np

df = pd.DataFrame(pd.read_csv('parsed_wine_ratings.csv'))
df
```

```python
df.vintage=df.vintage.astype('str')
```

```python
df.apply(lambda x: x.astype('str'))
```

```python
vintages = df['vintage']
```

```python
parsed_vintages = []
for vintage in vintages:
    if len(vintage) < 4:
        parsed_vintages.append(np.nan)
    elif len(vintage) > 4:
        parsed_vintages.append(vintage[:4])
    elif vintage in ['10.0', '12.0', '331.']:
        parsed_vintages.append(np.nan)
    else:
        parsed_vintages.append(vintage)
```

```python
df.drop('vintage', axis=1, inplace=True)
```

```python
df['vintage'] = parsed_vintages
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
```
